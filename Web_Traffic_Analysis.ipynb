5. How does the clickrate distribute across different links?

```python
# need dataframe grouped by linkid with clickrate aggregation
# drop preview events
# then get relative frequency of remaining event

df['event'].unique()
```

```python
clickrate_distribution_df = df[['event','linkid']]
clickrate_distribution_df
```

```python
# drop previews
clickrate_distribution_df = clickrate_distribution_df[clickrate_distribution_df['event'] != 'preview']
clickrate_distribution_df
```

```python
grouped_clicks_df = pd.DataFrame(clickrate_distribution_df.groupby('event').value_counts()['click'])
grouped_clicks_df
```

```python
grouped_clicks_df.reset_index(inplace=True)
grouped_clicks_df
```

```python
grouped_clicks_df.rename(columns={0: "clicks"}, inplace=True)
grouped_clicks_df
```

```python
grouped_pageview_df = pd.DataFrame(clickrate_distribution_df.groupby('event').value_counts()['pageview'])
grouped_pageview_df
```

```python
grouped_pageview_df.reset_index(inplace=True)
grouped_pageview_df
```

```python
grouped_pageview_df.rename(columns={0: "pageviews"}, inplace=True)
grouped_pageview_df
```

```python
merged_df = grouped_clicks_df.merge(grouped_pageview_df, on='linkid')
merged_df
```

```python
merged_df['clickrate']= merged_df['clicks'] / merged_df['pageviews']
merged_df
```

```python
grouped_clickrate_df = pd.DataFrame(merged_df.groupby('linkid')['clickrate'].sum())
grouped_clickrate_df
```

```python
grouped_clickrate_df.reset_index(inplace=True)
grouped_clickrate_df
```

```python
df_q6 = df[df['event'].isin(["click", "preview"])]
df_q6
```

```python
link_clicks = df[df["event"] == "click"].groupby(by='linkid')["event"].count().reset_index()
link_clicks
```

```python
link_clicks = link_clicks.rename(columns={"event": "clicks"})
link_clicks
```

```python
previews = df[df["event"] == "preview"].groupby(by='linkid')["event"].count().reset_index()
previews
```

```python
previews = previews.rename(columns={"event": "previews"})
previews
```

```python
# merge each into df_q6
df_q6 = pd.merge(left = df_q6, right=link_clicks, how='left', on='linkid')
df_q6 = pd.merge(left = df_q6, right=previews, how='left', on='linkid')
```

```python
df_q6.dropna(subset=["clicks", "previews"], inplace=True)
```
